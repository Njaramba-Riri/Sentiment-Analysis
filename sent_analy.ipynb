#### Data Preprocessing

```python
def clean_text(text):
    '''
    Make text lowercase, remove text in square brackets,remove links,remove punctuation, stopwords and remove words containing numbers.
    '''
    text = str(text).lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)

    stop_words=set(stopwords.words("english"))
    text = ' '.join(word for word in text.split(' ') if word not in stop_words)
    # only removing the hash # sign from the word
    text = re.sub(r'#', '', str(text)) 
    # remove punctuation
    punct = set(string.punctuation)
    text = "".join(ch for ch in text if ch not in punct) 
    " ".join(word for word in text.split() if word not in stop_words)
    
    return text
```

```python
full['clean_tweet']=full['text'].apply(clean_text)
full.head()
```

### Building the Models

##### vectorize the tweets

```python
#Count Vectorizer
cv=CountVectorizer( binary=True)#idicates if the word is present or not
cv.fit(train['clean_tweet'].values.tolist()+test['clean_tweet'].values.tolist())
train_cv=cv.transform(train['clean_tweet'])

#tfidf vectorizer
tf=TfidfVectorizer()
tf.fit(train['clean_tweet'].values.tolist()+test['clean_tweet'].values.tolist())
train_tf=tf.transform(train['clean_tweet'])
```

```python
#identify the target variable
y=train['sentiment']
```

```python
print(train_tf.shape, y.shape)
```

- Training the models

```python
from sklearn.model_selection import StratifiedKFold

folld=StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
```

```python
models={
    "SVM":LinearSVC(),
    "Logistic_Regression": LogisticRegression(),
    "Naive_Bayes":BernoulliNB()
}

#vectors
vectors={
    "CountVectorizer":train_cv,
    "TfidFVectorizer":train_tf
}
```

```python
"""DO a cross-validation(KFold)"""
def strat_kfold(class_model:str,sample:str,y,kfold):
    
    results={}
    
    #store the model names in a dict
    results['model name']=class_model+ "_" + sample
    
    #call the model and the training data
    model=models[class_model]
    X= vectors[sample]
    
    #perform the cross_validation
    for fold, (train_idx, valid_idx) in enumerate(kfold.split(X,y)):
        X_train, X_valid=X[train_idx], X[valid_idx]
        y_train, y_valid=y[train_idx], y[valid_idx]
        
        #train and predict
        model.fit(X_train, y_train)
        y_pred=model.predict(X_valid)
        
        
        results["Kfold_{}".format(fold+1)]=f1_score(y_valid, y_pred)
    return results
```

```python
#storing the models
all_models=[]

for class_model in models:
    for sample in vectors:
        all_models.append(strat_kfold(class_model, sample, y, folld))
        print(f"Model: {class_model}_{sample}...\n")
```

```python
models_df=pd.DataFrame(all_models)
models_df
```

#### XGBoost Model

```python
params={
    "objective":'binary:logistic',
    "tree_method":'hist',
}
```

```python
xg_dict={}
xg_dict['model name']="XGBoost_CV"


for fold, (train_idx, valid_idx) in enumerate(folld.split(train_cv, y)):
    print(f"\nCurrently Training: {xg_dict['model name']}... Fold: {fold+1}")
    X_train, X_valid= train_tf[train_idx].astype(np.float32), train_tf[valid_idx].astype(np.float32)
    y_train, y_valid= y[train_idx].astype(np.float32), y[valid_idx].astype(np.float32)

    #training
    xgb_model=xgb.XGBClassifier(**params)
    xgb_model.fit(
        X_train,y_train, 
        eval_set=[(X_valid, y_valid)], 
        verbose=100,
        early_stopping_rounds=50
        )
    
    #predictions
    y_preds=xgb_model.predict(X_valid)
    xg_dict["Kfold_{}".format(fold+1)]=f1_score(y_valid, y_preds)

```

```python
# adding results to models df
new_model = pd.DataFrame(xg_dict, columns=models_df.columns, index=[0])
models_df = pd.concat([models_df, new_model], ignore_index=True)
```

```python
xgb_dict={}
xgb_dict['model name']="XGBoost_tfidf"

for fold, (train_idx, valid_idx) in enumerate(folld.split(train_tf, y)):
    print(f"\nCurrently Training: {xgb_dict['model name']}... Fold: {fold+1}")
    X_train, X_valid= train_tf[train_idx].astype(np.float32), train_tf[valid_idx].astype(np.float32)
    y_train, y_valid= y[train_idx].astype(np.float32), y[valid_idx].astype(np.float32)


    #training
    xgb_model=xgb.XGBClassifier(**params)
    xgb_model.fit(
        X_train, y_train, 
        eval_set=[(X_valid, y_valid)],
        eval_metric= ['logloss'], 
        verbose=1000, 
        early_stopping_rounds=50
        )
    

    #predictions
    y_preds=xgb_model.predict(X_valid)
    xgb_dict['Kfold_{}'.format(fold+1)]=f1_score(y_valid, y_preds)


```

```python
# adding results to models df
new_model = pd.DataFrame(xgb_dict, columns=models_df.columns, index=[0])
models_df = pd.concat([models_df, new_model], ignore_index=True)
```
